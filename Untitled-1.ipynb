## Mysql + concurrent

```python
import pymysql
from concurrent.futures import ThreadPoolExecutor
from sqlalchemy import create_engine
import pandas as pd
```

```python
# 设置MySQL连接
connection = pymysql.connect(
    host='localhost',
    user='root',
)
```

```python
with connection.cursor() as cursor:
    cursor.execute("CREATE DATABASE IF NOT EXISTS comments;")
    print("Database created successfully.")
```

```python
database_name = 'comments'  # 你的数据库名称
engine = create_engine(f"mysql+pymysql://root@localhost/{database_name}")
```

```python
# 读取CSV文件并分块
def load_data_in_chunks(file_path, chunk_size):
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        yield chunk

# 定义插入数据块的函数
def insert_chunk_to_db(chunk, table_name):
    chunk.to_sql(table_name, con=engine, if_exists='append', index=False)

# 多线程并行处理
def parallel_insert(file_path, table_name, chunk_size=20000, num_threads=6):
    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        for chunk in load_data_in_chunks(file_path, chunk_size):
            executor.submit(insert_chunk_to_db, chunk, table_name)

# 运行并行插入
file_path = '/Users/fengcharles/Desktop/8083Porj/comments data/test.csv'
parallel_insert(file_path, test)
print("Data loaded successfully.")
```
